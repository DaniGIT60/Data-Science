```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample Tesla stock data
data = {
    'Date': pd.date_range(start='2024-01-01', periods=30, freq='D'),
    'Close': [1250 + i * 5 + (i % 3) * 10 for i in range(30)]  # Hypothetical stock prices
}

# Create a DataFrame
tesla_stock_data = pd.DataFrame(data)

# Plotting Tesla stock data
plt.figure(figsize=(10, 6))
plt.plot(tesla_stock_data['Date'], tesla_stock_data['Close'], label='Tesla Closing Price')
plt.title("Tesla Stock Price Over Time")
plt.xlabel("Date")
plt.ylabel("Stock Price (USD)")
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample GameStop stock data
data_gamestop = {
    'Date': pd.date_range(start='2024-01-01', periods=30, freq='D'),
    'Close': [150 + i * 2 - (i % 4) * 3 for i in range(30)]  # Hypothetical stock prices
}

# Create a DataFrame
gamestop_stock_data = pd.DataFrame(data_gamestop)

# Plotting GameStop stock data
plt.figure(figsize=(10, 6))
plt.plot(gamestop_stock_data['Date'], gamestop_stock_data['Close'], label='GameStop Closing Price', color='red')
plt.title("GameStop Stock Price Over Time")
plt.xlabel("Date")
plt.ylabel("Stock Price (USD)")
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
